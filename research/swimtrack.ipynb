```python
import supervision as sv
from ultralytics import YOLO

MODEL = "yolov8x.pt"
model = YOLO(MODEL)
byte_tracker = sv.ByteTrack()
annotator = sv.BoxAnnotator()
```

```python
import numpy as np

def process_frame(frame: np.ndarray, index: int) -> np.ndarray:
    results = model(frame)[0]
    detections = sv.Detections.from_ultralytics(results)
    detections = byte_tracker.update_with_detections(detections)
    labels = [
        f"#{tracker_id} {model.model.names[class_id]} {confidence:0.2f}"
        for _, _, confidence, class_id, tracker_id
        in detections
    ]
    return annotator.annotate(scene=frame.copy(), detections=detections, labels=labels)
```

```python
from pathlib import Path

CWD = Path().absolute()
# VIDEO_PATH = CWD / "istockphoto-1371542874-640_adpp_is.mp4"
VIDEO_PATH = CWD / "single.mp4"

sv.process_video(source_path=str(VIDEO_PATH), target_path=f"result_single.mp4", callback=process_frame)
```
